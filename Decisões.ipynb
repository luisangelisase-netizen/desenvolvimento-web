# Decisões

Estruturas que permitem alterar o fluxo de execução do programa de acordo com uma ou mais condições.

```python

  bloco verdadeiro
```

**Atenção** com a identação.

**Identação:** Organização do código usando espaços ou tabulações para alinhar blocos e indicar hierarquia.

**Exemplo:** Ler um número. Caso for o número 1, mostre uma mensagem.

**Exemplo:** Os blocos de identação devem ter sempre o mesmo tamanho (quantidadade de espaços).

````python
i = int(input("Digite um número:"))
if i == 1:
  #Cuidado com a identação
  print("Você digitou o número 1.")
  print("Segunda mensagem dentro do IF.")
  # o print executa depois do if
  #a identação está no memso nivel
print("FIM depois da decisão")
````

````python
i = int(input("Digite um número:"))
if i == 1:
  # executa este bloco
  # somente se a decisão for verdadeira
  print("Você digitou o número 1.")
  # o print executa depois do if
  #a identação está no memso nivel
print("FIM depois da decisão")
````
